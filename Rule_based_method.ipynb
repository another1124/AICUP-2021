# Extract data from ZIP file

```python
import pandas as pd
import numpy as np
import zipfile
import os
import re

#Extract data from ZIP file
def ZipFileProcessing (ZipfilePath):
    ids=[]
    txtlist=[]
        
    if zipfile.is_zipfile(ZipfilePath): 
        zip = zipfile.ZipFile(ZipfilePath, 'r')
        for files in zip.namelist():
            if '.txt' in files:
                zfile = zip.read(files)
                ids.append(int(re.sub("[^0-9]", "", files)))
                txtlist.append(zfile.decode())
        zip.close()
    data ={'Id': ids ,'Sentence': txtlist }
    df = pd.DataFrame(data)

    return df
```

```python
df = ZipFileProcessing('./gdrive/MyDrive/machine/final/data/Train.zip')
#Sort dataframe
df = df.sort_values(by = ['Id'])
df = df.reset_index(drop=True)
df.head()
```

# Rule Based Method




```python
def get_key_list(sentence , keyword_db):
    
    keyword_id = []
    for index in range(len(keyword_db)):
        #keyword_id.append(index)
        count = 0
        for obj in keyword_db.iloc[index].values:
            if obj is not np.nan :
               count = count + sentence.count(obj)
        if count>0:
            keyword_id.append(index)
    return keyword_id
```

```python
key_pest_ids_list = []
key_crop_ids_list = []
key_chem_ids_list = []
key_season_ids_list = []
key_local_ids_list = []
#key_localold_ids_list = []
for sentence in test_df['Sentence']:#//test_df or df
    key_pest_ids_list.append(get_key_list(sentence, df_key_pest))
    key_crop_ids_list.append(get_key_list(sentence, df_key_crop))
    key_chem_ids_list.append(get_key_list(sentence, df_key_chem))
    key_season_ids_list.append(get_key_list(sentence, df_key_season))
    key_local_ids_list.append(get_key_list(sentence, df_key_local))
    #key_localold_ids_list.append(get_key_list(sentence, df_key_localold))
```

```python
from tqdm.notebook import tqdm 

chem_score_list = []
crop_score_list = []
pest_score_list = []
season_score_list = []
local_score_list = []
#localold_score_list = []
include_score_list = []
id_test_list = []
id_reference_list = []
#elem_score_list = []
Ids = df_key_ids['ID']
for i in tqdm(range(len(df_key_ids))):
    id_test = Ids[i]
    sen_1_chem = df_key_ids['key_chem_ids'][i]
    sen_1_crop = df_key_ids['key_crop_ids'][i]  
    sen_1_pest = df_key_ids['key_pest_ids'][i]
    sen_1_season = df_key_ids["key_season_ids"][i]
    sen_1_local = df_key_ids["key_local_ids"][i]
   # sen_1_localold = df_key_ids["key_localold_ids"][i]
    sen_1_elem = len(sen_1_chem)+len(sen_1_crop)+len(sen_1_pest)
    for j in range(len(df_key_ids)):
        id_reference = Ids[j]
        if id_test == id_reference:
                continue
        sen_2_chem = df_key_ids['key_chem_ids'][j]
        sen_2_crop = df_key_ids['key_crop_ids'][j]  
        sen_2_pest = df_key_ids['key_pest_ids'][j]
        sen_2_season = df_key_ids["key_season_ids"][j]
        sen_2_local = df_key_ids["key_local_ids"][j]
       # sen_2_localold = df_key_ids["key_localold_ids"][j]
        sen_2_elem = len(sen_2_chem)+len(sen_2_crop)+len(sen_2_pest)
        if len(sen_1_chem)!=0:
            chem = len(set(sen_1_chem).intersection(sen_2_chem))/len(sen_1_chem)#(len(set(sen_1_chem).union(sen_2_chem)))
        else: 
            chem = 0
        if len(sen_1_crop)!=0:
            crop = len(set(sen_1_crop).intersection(sen_2_crop))/len(sen_1_crop)#(len(set(sen_1_crop).union(sen_2_crop)))
        else:
            crop = 0
        if len(sen_1_pest)!=0:
            pest = len(set(sen_1_pest).intersection(sen_2_pest))/len(sen_1_pest)#(len(set(sen_1_pest).union(sen_2_pest)))
        else:
            pest = 0
        if  len(sen_1_season)!=0 and len(sen_2_season)!=0:
            season = len(set(sen_1_season).intersection(sen_2_season))/(len(set(sen_1_season)))
        else: 
            season = 0
        if  len(sen_1_local)!=0 and len(sen_2_local)!=0:

            
            local = len(set(sen_1_local).intersection(sen_2_local))/(len(set(sen_1_local)))
            if local == 0:
              local = -1
        else: 
            local = 0
       # if  len(sen_1_localold)!=0 and len(sen_2_localold)!=0:
       #
       #     
       #     localold = len(set(sen_1_localold).intersection(sen_2_localold))/(len(set(sen_1_localold)))
       #     if localold == 0:
       #       localold = -1
       # else: 
       #     localold = 0
        if sen_2_elem!=0:
          inflag=sen_1_elem/sen_2_elem
        else:
          inflag=100
        chem_score_list.append(chem)
        crop_score_list.append(crop)
        pest_score_list.append(pest)
        season_score_list.append(season)
        local_score_list.append(local)
        include_score_list.append(inflag)
       # localold_score_list.append(localold)
        id_test_list.append(id_test)
        id_reference_list.append(id_reference)
```

```python
data ={'Test': id_test_list ,'Reference': id_reference_list ,'chem_score': chem_score_list, \
       'crop_score': crop_score_list, 'pest_score':pest_score_list ,"season_score":season_score_list,"local_score":local_score_list,"include_score":include_score_list}#,"localold_score":localold_score_list
score_list = pd.DataFrame(data)

score_list
```
